- Position/Players Vs Hitting

- Position/Players Vs Serving

# Importing Libraries

```python
import numpy as np # Linear Algebra
import pandas as pd # File Proccesing (Read CSV files)
import matplotlib.pyplot as plt # Data Visualization
import seaborn as sns         
```

# Reading the Data 

```python
df = pd.read_csv('VNL2023.csv') # Read in the CSV
```

# Checking Data and Looking for Missing Values 

```python
df.head() # First 5 rows
```

```python
df.shape # (rows, cols)
```

```python
df.describe()
```

```python
sns.heatmap(df.isnull()) # No missing values ✅
```

# Distribution of the Ages of the Players

```python
df_temp=df['Age'].sort_values().head(150)

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.countplot(x=df_temp, palette='viridis')

plt.xlabel('Player Age')
plt.ylabel('Count')
plt.title('Player Age Distribution')

plt.show()
```

# Position Distribution

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,6))
sns.countplot(x=df['Position'], palette='hls')

plt.xlabel('Positions')
plt.ylabel('Count')
plt.title('Position Distribution')

plt.show()
```

# Position/Players Vs Hitting

```python
sns.set(style='whitegrid')
plt.figure(figsize=(30, 15))
sns.barplot(x=df['Player'], y=df['Attack'], palette='Blues_d')

plt.xlabel('Player Names')
plt.ylabel('Attack')
plt.title('# of Attacks by Player')
plt.xticks(rotation=90) # rotate names by 90 degrees 

plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.barplot(x=df['Position'], y=df['Attack'], palette='magma')

plt.xlabel('Positions')
plt.ylabel('Attack')
plt.title('# of Attacks by Position')

plt.show()
```

# Position/Players Vs Passing (Dig & Receive)

```python
sns.set(style='whitegrid')
plt.figure(figsize=(30, 15))
sns.barplot(x=df['Player'], y=df['Dig'], palette='mako')

plt.xlabel('Player Names')
plt.ylabel('# of Digs')
plt.title('# of Digs by Player')
plt.xticks(rotation=90) # rotate names by 90 degrees 

plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.barplot(x=df['Position'], y=df['Dig'], palette='crest')

plt.xlabel('Positions')
plt.ylabel('# of Digs')
plt.title('# of Digs by Position')

plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(30, 15))
sns.barplot(x=df['Player'], y=df['Receive'], palette='flare')

plt.xlabel('Player Names')
plt.ylabel('# of Receives')
plt.title('# of Reveives by Player')
plt.xticks(rotation=90) # rotate names by 90 degrees 

plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.barplot(x=df['Position'], y=df['Receive'], palette='rocket')

plt.xlabel('Positions')
plt.ylabel('# of Receives')
plt.title('# of Receives by Position')

plt.show()
```
